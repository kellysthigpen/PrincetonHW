```python
#3D box visualization of chicken over time
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Load CSV
df = pd.read_csv("ChickenGrowth.csv")

# Function to draw a box given length, width, height
def draw_box(ax, L, W, H, color="skyblue", alpha=0.3):
    # Define the 8 corners of the box
    x = [0, L, L, 0, 0, L, L, 0]
    y = [0, 0, W, W, 0, 0, W, W]
    z = [0, 0, 0, 0, H, H, H, H]

    # Define the 6 faces by grouping corners
    vertices = [
        [ [x[0],y[0],z[0]], [x[1],y[1],z[1]], [x[2],y[2],z[2]], [x[3],y[3],z[3]] ], # bottom
        [ [x[4],y[4],z[4]], [x[5],y[5],z[5]], [x[6],y[6],z[6]], [x[7],y[7],z[7]] ], # top
        [ [x[0],y[0],z[0]], [x[1],y[1],z[1]], [x[5],y[5],z[5]], [x[4],y[4],z[4]] ], # front
        [ [x[2],y[2],z[2]], [x[3],y[3],z[3]], [x[7],y[7],z[7]], [x[6],y[6],z[6]] ], # back
        [ [x[1],y[1],z[1]], [x[2],y[2],z[2]], [x[6],y[6],z[6]], [x[5],y[5],z[5]] ], # right
        [ [x[0],y[0],z[0]], [x[3],y[3],z[3]], [x[7],y[7],z[7]], [x[4],y[4],z[4]] ], # left
    ]

    ax.add_collection3d(Poly3DCollection(vertices, facecolors=color, linewidths=1, edgecolors="black", alpha=alpha))

# Example: draw chicken at row 3
row = df.iloc[5]
L, W, H = row["length(cm)"] * 1.5, row["width(cm)"] * 1.5, row["height(cm)"] * 1.5

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

draw_box(ax, L, W, H)

ax.set_xlabel("Length (cm)")
ax.set_ylabel("Width (cm)")
ax.set_zlabel("Height (cm)")
ax.set_title(f"Chicken at time {row['time (years)']} years")
plt.show()
```
