```python
# Fit a linear model to predict porosity from grain size using matrix algebra (no built-in regression functions).
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(0)
df = pd.DataFrame({
    "grain_size": np.random.normal(0.5, 0.1, 100),
    "porosity": np.random.normal(0.3, 0.05, 100),
    "depth": np.linspace(0, 1000, 100)
})

df.head(10)

X = np.array([[1,0.676405,0],
             [1,0.540016,10.101001],
             [1,0.597874,20.202020]])
y = np.array([0.394158,
              0.232612,
              0.236476])

# --- Solve using matrix algebra: β = (XᵀX)⁻¹Xᵀy ---
beta = np.linalg.inv(X.T @ X) @ X.T @ y
print(beta)

#plt.scatter(df["grain_size"],df["porosity"], color = "b",zorder=3,alpha=0.5)

#plt.xlabel("Grain Size")
#plt.ylabel("Porosity")
#plt.axis('equal')
#plt.title("Predicting Porosity\nFrom Grain Size")

```
